```python
# create new csv file to save all the collected links in it
csv_file = open('wuzzuf_job_links_20-12-2022.csv', 'a+', encoding="utf-16")
writer = csv.DictWriter(csv_file, fieldnames=['title', 'link'])
writer.writeheader()
```

```python
rows = 1 # to know number of rows for my data
for url in MAIN_URLS:
    # to view progress while looping
    print('-'*50)
    print(f'start URL: {url}')
    print(f'start row: {rows}')
    
    # start new category scrapping so start page is 0
    page = 0
    time.sleep(30)
    # loop through all jobs in the category
    while(True):
        try:
            # to view progress while looping
            print(f'Page: {page}')

            # send request to the website to collect data
            link = url + str(page)
            headers = {
                'User-Agent': random.choice(USER_AGENTS),
                "Accept-Encoding": "*",
                "Connection": "keep-alive"
            }
            response = requests.get(link, headers=headers, timeout=100)

            # collect data and save it to csv file
            soup =  BeautifulSoup(response.text, 'html.parser')
            jobs = soup.find_all('div', class_='css-1gatmva e1v1l3u10')
            for job in jobs:
                rows += 1
                writer.writerow({
                     'title': job.find('h2', class_='css-m604qf').find('a').text,
                     'link': f"https://wuzzuf.net{job.find('h2', class_='css-m604qf').find('a').get('href')}"
                    })

            # when no more jobs break out of the loop
            if len(jobs) <= 0:
                break

            # go to the next page by increment the page
            page += 1

            time.sleep(5)
        except Exception as e:
            print(f'Retrying page: {page}')
    print('-'*50)
csv_file.close()
```
